```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('Vehicle_Population.csv')
pop2  = pop[(pop['Fuel Type'] =='Electric') & (pop['Make'] == 'Tesla')]
pop2
```

```python
pop3 = pop2.groupby(["County","Data Year"]).sum().sort_values(by = 'Number of Vehicles', ascending = False)
pop3 = pop3.reset_index(inplace = False)
```

```python
socal = ['Los Angeles','Orange','San Diego','Riverside','San Bernardino','Ventura','Santa Barbara','Kern','San Luis Obispo','Imperial']
```

```python
pop4 = pop3[pop3['County'].isin(socal)]
pop4
```
